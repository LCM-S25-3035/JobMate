```python
from jobspy import scrape_jobs
```

```python
import pandas as pd
```

```python
jobList = pd.read_csv('Data-Related_Job_Titles.csv')
```

```python
jobList
```

```python
for keyWord in jobList['Data & Business Analytics Jobs']:
    print(keyWord)
```

```python
all_jobs = []
for keyWord in jobList['Data & Business Analytics Jobs']:
    jobs = scrape_jobs(
        site_name=["indeed"],
        search_term=keyWord,
        location="Canada",
        results_wanted=100,
        hours_old=200,
        country_indeed="canada" 
    )
    print(f"{keyWord}: Found {len(jobs)} jobs.")
    all_jobs.append(jobs)
```

```python
combined_jobs = pd.concat(all_jobs, ignore_index=True)
```

```python
combined_jobs
```

```python
combined_jobs.to_csv("data_jobs_indeed.csv", index=False)
```
